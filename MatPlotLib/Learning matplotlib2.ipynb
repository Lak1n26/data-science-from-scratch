```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

### Линейные графики
y = f(x)

```python
fig = plt.figure()
ax = plt.axes()
```

```python
x = np.linspace(0, 10, 1000)
plt.style.use('seaborn-whitegrid')
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

### Настройка графика: цвета и стили линий 

```python
plt.plot(x, np.sin(x), color = 'chartreuse')
# color =
# 'green'
# 'g' - (rgbcmyk) = red, green, blue, cyan, magenta, yellow, blacK
# '0' - gray color, '1' - white
# (1, 0 , 0) - RGB
# '#212121' - RRGGBB от 00 до FF
```

```python
plt.plot(x, np.sin(x), linestyle = ':')
# solid (-)
# dashed - - - - - - - (--)
# dashdot -.-.-.-.-.-.- (-.)
# dotted ............ (:)
```

```python
plt.plot(x, x, '-g')
plt.plot(x, x + 1, '--r')
plt.plot(x, x + 2, ':b')
plt.plot(x, x + 3, '-.y')
```

### Пределы осей координат 

```python
x = np.linspace(0, 10, 1000)
y = np.sin(x)
plt.plot(x, y)
plt.xlim(10, 0)
plt.ylim(1, -1)
```

```python
plt.plot(x, y, '-g')
plt.axis('equal')
```
