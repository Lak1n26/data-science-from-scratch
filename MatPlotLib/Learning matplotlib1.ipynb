```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
```

```python
years = np.arange(2016, 2023)
gdp = np.array([300, 540, 1075, 2862, 5979, 10289, 14958])
```

```python
fig = plt.figure()
plt.plot(years, gdp, color='green', marker='o')
plt.title("ВВП")
plt.xlabel('Годы')
plt.ylabel('Трлн $')
fig.savefig("GDP by 2016-2022 years")
```

### Столбчатые графики

```python
subjects = np.array(['Math', 'Physics', 'PE', 'Literature'])
points = np.array([10, 2, 9, 4])
plt.bar(range(subjects.size), points)
plt.title("Предметы")
plt.ylabel("Количество очков")
plt.xticks(range(subjects.size), subjects)
```

```python
from collections import Counter
grades = np.random.randint(0, 100, 14)

histogram = Counter(min(grade // 10 * 10 + 10, 90) for grade in grades)
print(histogram)
```

```python
plt.bar([x + 5 for x in histogram.keys()], histogram.values(), 10, edgecolor=(0, 0, 0))
plt.xticks([10 * x for x in range(11)])
plt.xlabel("Дециль")
plt.ylabel("Количество студентов")
plt.title("Оценки за экзамен")
plt.axis([-5, 105, 0, 5])
```

## Задача по статистике
Дано N = 30.
Необходимо разбить N на интервалы и построить гистограмму

```python
n = 30
bins = 1 + 3.322 * np.log10(n)
# bins = 5.9 -> (количество интервалов 5 или 6)
bins = 5
x = np.random.randint(45, 96, 30)
x
```

```python
bin_len = (x.max() - x.min()) // bins
# bin_len = 10
from collections import Counter
intervals = Counter(min(param // bin_len * bin_len + (bin_len // 2), x.max()-bin_len) for param in x)
print(intervals)
plt.bar(intervals.keys(), intervals.values(), bin_len, edgecolor=(0,0,0))
plt.axis([30, 100, 0, 8])
```
