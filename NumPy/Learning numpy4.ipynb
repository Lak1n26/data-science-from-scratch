```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

### "Прихотливая" индексация (fancy indexing)

```python
x = np.random.randint(0, 20, size = (3, 4))
row = np.array([0,1])
col = np.array([0, 2])
x
```

```python
x[row, col] # возвращает элементы с индексами (0,0) и (1, 2)
```

```python
# комбинированная индексация
x[1, [3, 0, 2]]
x[1:, [3, 2, 1]]
```

```python
mean = [0, 0]
cov = [[1, 2], [2, 5]]
x = np.random.multivariate_normal(mean, cov, 50)
print(x.shape)

seaborn.set()
indices = np.random.choice(x.shape[0], 20, replace = False)
selection = x[indices]

plt.scatter(x[:, 0], x[:, 1], alpha = 0.3)
plt.scatter(selection[:, 0], selection[:, 1], facecolor = 'red', s = 100, alpha = 0.3)
```
